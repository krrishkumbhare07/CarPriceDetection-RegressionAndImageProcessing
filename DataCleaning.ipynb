```python
import pandas as pd
import numpy as np

data=pd.read_csv("car_details_cardekho.csv")

'''
Data Cleaning

1. Removing duplicate values
2. Replacing values filled "NaN" with 0 or mean value
3. Finding mean
4. Creating new dataframe
'''

print(data.head())
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates(subset=None,inplace=True)
data.info()
```

```python
for i in range(len(data["year"])):
    if type(data.iloc[i]["year"])!=np.int64:
        data=data.drop(data.index[i])  
      
```

```python
data=data.dropna()
data
```

```python
for i in range(len(data["year"])):
    if data.iloc[i]["year"]>np.int64(2023) or data.iloc[i]["year"]<np.int64(2007):
        data.drop(data.index[i])
```

```python
data.to_csv("Car_price_data_cleaned.csv",index=False)
```
